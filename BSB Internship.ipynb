```python
import matplotlib.pyplot as plt
```

```python
# Line Plot
plt.plot(["Finland","Denmark","Iceland","Sweden","Israel"], ["Regional indicator","Ladder score","Log GDP per capita","Social support","Healthy life expectancy",])
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Basic Line Plot")
plt.figure(figsize = (5,3))
```

```python
# Bar Plot
x = ['Finland','Denmark','Iceland','Sweden','Israel']
y =  [0.695, 0.699, 0.718, 0.724,	0.740]	
plt.barh(x,y, color='Black')  #h represents the horizontal graph
```

```python
# Scatter Plot
import matplotlib.pyplot as plt
x = ['Denmark','Iceland','Sweden']
y = ['Regional Indicator','Ladder score','Log GDP per capita']
plt.scatter(x,y)
```

```python
# Histogram
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for the histogram
data = np.random.randn(1000)

# Plotting a basic histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black')

# Adding labels and title
plt.xlabel('Country')
plt.ylabel('Regional indicator')
plt.title('Basic Histogram')

# Display the plot
plt.show()
```

```python
import seaborn as sns


sns.set(style="dark")
fmri = sns.load_dataset("fmri")

# Plot the responses for different\
# events and regions
sns.lineplot(x="timepoint",
             y="signal",
             hue="region",
             style="event",
             data=fmri)
```

```python
import numpy as np
import seaborn as sns

sns.set(style="white")

# Generate a random univariate dataset
rs = np.random.RandomState(10)
d = rs.normal(size=100)

# Plot a simple histogram and kde
sns.histplot(d, kde=True, color="m")
```

```python
import seaborn as sns

sns.set(style="ticks")

# Loading the dataset
df = sns.load_dataset("anscombe")

# Show the results of a linear regression
sns.lmplot(x="x", y="y", data=df)
```

```python
# importing packages 
import seaborn as sns 
import matplotlib.pyplot as plt 

# loading dataset 
data = sns.load_dataset("iris") 


def plot():
    sns.lineplot(x="sepal_length", y="sepal_width", data=data) 

with sns.axes_style('darkgrid'):
    
    # Adding the subplot
    plt.subplot(211)
    plot()
    
plt.subplot(212)
plot()
```
